```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
filepath='https://raw.githubusercontent.com/allisonhorst/palmerpenguins/master/inst/extdata/penguins.csv'
pinguins = pd.read_csv(filepath)
pinguins.head()
```

```python
pinguins['sex']
```

```python
data_female=pinguins[pinguins['sex']=='female']
data_female
```

```python
print(len(pinguins))
pinguins.dropna(inplace=True)
len(pinguins)
```

```python
print(len(pinguins))
pinguins.drop_duplicates(inplace=True)
len(pinguins)
```

```python

body_mass_g = pinguins['body_mass_g']
mean_mass = body_mass_g.mean()
std_mass = body_mass_g.std()
z_score = (body_mass_g - mean_mass) / std_mass
z_score


treshold=2.5

outliers = pinguins[np.abs(z_score) > treshold]
outliers
```
